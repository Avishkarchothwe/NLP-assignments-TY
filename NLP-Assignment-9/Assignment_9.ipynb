3. Synonymy (Synonyms)

Synonyms are words that have similar meanings.

```python
word = "happy"

synonyms = set()

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms:", synonyms)
```

4. Antonymy (Antonyms)

Antonyms are words with opposite meanings.

```python
word = "happy"

antonyms = set()

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Antonyms:", antonyms)
```

5. Hypernymy (Hypernyms)

Hypernyms are more general categories of a word.

```python
word = "dog"

hypernyms = set()

for syn in wn.synsets(word):
    for hyper in syn.hypernyms():
        hypernyms.add(hyper.name())

print("Hypernyms:", hypernyms)
```

6. Extract Relationships from Text

```python
from nltk.corpus import wordnet as wn

word = "car"

synonyms = []

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())

print("Synonyms:", set(synonyms))
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn

nltk.download('punkt') # Ensure the main 'punkt' tokenizer is downloaded
nltk.download('punkt_tab') # Explicitly download 'punkt_tab' as suggested by the error

text = "The dog chased the cat in the garden"

words = word_tokenize(text)

for word in words:

    synsets = wn.synsets(word)

    if synsets:
        syn = synsets[0]

        synonyms = [l.name() for l in syn.lemmas()]

        antonyms = []
        for l in syn.lemmas():
            if l.antonyms():
                antonyms.append(l.antonyms()[0].name())

        hypernyms = [h.name() for h in syn.hypernyms()]

        print("\nWord:", word)
        print("Synonyms:", synonyms)
        print("Antonyms:", antonyms)
        print("Hypernyms:", hypernyms)
```
