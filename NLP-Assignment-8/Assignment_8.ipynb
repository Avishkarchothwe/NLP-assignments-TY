```python
import spacy
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
data = [
("Apple CEO Tim Cook visited India.",
 {"entities":[("Apple","ORG"),("Tim Cook","PERSON"),("India","GPE")]}),

("Microsoft was founded by Bill Gates.",
 {"entities":[("Microsoft","ORG"),("Bill Gates","PERSON")]}),

("Narendra Modi met Joe Biden in Washington.",
 {"entities":[("Narendra Modi","PERSON"),("Joe Biden","PERSON"),("Washington","GPE")]})
]
```

```python
predicted_labels = []
true_labels = []

for text, annotation in data:

    doc = nlp(text)

    predicted = [(ent.text, ent.label_) for ent in doc.ents]
    actual = annotation["entities"]

    print("\nSentence:", text)
    print("Predicted:", predicted)
    print("Actual:", actual)

    predicted_labels.extend([label for _,label in predicted])
    true_labels.extend([label for _,label in actual])
```

```python
min_len = min(len(predicted_labels), len(true_labels))

predicted_labels = predicted_labels[:min_len]
true_labels = true_labels[:min_len]
```

```python
accuracy = accuracy_score(true_labels, predicted_labels)
print("Accuracy:", accuracy)
```

```python
report = classification_report(true_labels, predicted_labels)
print(report)
```
