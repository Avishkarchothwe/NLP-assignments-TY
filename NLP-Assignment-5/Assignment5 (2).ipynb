```python
# Import required libraries
import nltk
from nltk.tokenize import WhitespaceTokenizer, wordpunct_tokenize, TreebankWordTokenizer
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import MWETokenizer
from nltk.stem import PorterStemmer, SnowballStemmer
```

```python
# Download required resources (run once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Sample text
text = "Hello! I am learning Natural Language Processing using NLTK. I love machine learning and data science."

print("Original Text:\n", text)
```

```python
# -----------------------------
# TOKENIZATION
# -----------------------------

print("\n1. Whitespace Tokenization")
ws_tokenizer = WhitespaceTokenizer()
print(ws_tokenizer.tokenize(text))

print("\n2. Punctuation-based Tokenization")
print(wordpunct_tokenize(text))

print("\n3. Treebank Tokenization")
tree_tokenizer = TreebankWordTokenizer()
print(tree_tokenizer.tokenize(text))

print("\n4. Tweet Tokenization")
tweet_tokenizer = TweetTokenizer()
print(tweet_tokenizer.tokenize("Learning NLP is fun! #AI #MachineLearning :)"))

print("\n5. MWE Tokenization (Multi Word Expression)")
mwe_tokenizer = MWETokenizer([('machine', 'learning'), ('data', 'science')])
print(mwe_tokenizer.tokenize(text.split()))
```

```python
# -----------------------------
# STEMMING
# -----------------------------

words = ws_tokenizer.tokenize(text)

print("\nWords for Stemming:", words)
```

```python
# Snowball Stemmer
snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in words]

print("\nSnowball Stemmer Output:")
print(snowball_stems)
```

```python
from nltk.stem import WordNetLemmatizer

# Lemmatization
lemmatizer = WordNetLemmatizer()

lemmas = [lemmatizer.lemmatize(word) for word in words]

print("\nLemmatization Output:")
print(lemmas)
```
